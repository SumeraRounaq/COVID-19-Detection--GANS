# Getting positive Chest X-Ray images from Cohen repository

```python
import pandas as pd
import shutil
import os
from pathlib import Path
```

```python
# Selecting all combination of 'COVID-19' patients with 'PA' X-Ray view
virus = "Pneumonia/Viral/COVID-19" # Virus to look for
x_ray_view = "PA"   # View of X-Ray


covid_path = ".\covid-chestxray-dataset-0.41-Cohen\covid-chestxray-dataset-0.41"
csv = os.path.join(covid_path ,  'metadata.csv')
print(csv)
imageDir = os.path.join(covid_path, 'images')
print(imageDir)

directory = 'COVID19-PA'
parent_dir = Path().absolute()
print(parent_dir)
outputDir = os.path.join(parent_dir, directory)

os.mkdir(outputDir)
print("Directory '% s' created" % directory)


covid_df = pd.read_csv(csv)

covid_df[0:3]


    
    



```

```python
count=0

for(i, row) in covid_df.iterrows():
    if row["finding"] != virus or row["view"] != x_ray_view:
        continue
    filename = row["filename"].split(os.path.sep)[-1]
    
    inputPath = os.path.sep.join([imageDir,filename])
    print("Input Path:", inputPath)
    outputPath =  os.path.sep.join([outputDir,filename])
    print("Output Path:", outputPath)
    shutil.copy2(inputPath,outputPath)
    count += 1
    
        
    
```
